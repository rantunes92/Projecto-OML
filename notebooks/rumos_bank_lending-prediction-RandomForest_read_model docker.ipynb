Neste notebook vamos ler o modelo escolhido da Random Florest

```python
import mlflow
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
import joblib
import mlflow.sklearn
```

Definir a directoria onde as experiencias são guardadas

```python
from pathlib import Path

uri = "http://127.0.0.1:5000"



mlflow.set_tracking_uri(uri)
```

```python
# Fazer load da versão do modelo "Random_Forest" que registamos no notebook anterior
# Têm de mudar a versão de acordo com quantas versões registaram.

model_name = "Random_Forest"
model_version = "11"
f"models:/{model_name}/{model_version}"
```

```python
model = mlflow.sklearn.load_model(f"models:/{model_name}/{model_version}")
model
```

Vamos usar o modelo para fazer previsões

```python
data_path = '../data/rumos_bank_test.csv'
df = pd.read_csv(data_path)
```

```python
nsamples=10
input_data = df.loc[df['default.payment.next.month'] == 1].sample(nsamples)
input_data
```

```python
input = input_data.drop("default.payment.next.month", axis=1)
colunas=input.columns
input
```

```python
scaler = MinMaxScaler()
scaler = joblib.load("scaler.pkl")
input = scaler.transform(input)
input=pd.DataFrame(input, columns = colunas) 
```

```python
input.dtypes
```

```python
print(model)
```

```python
print(input.dtypes)
print(input.head())
```

```python
#model.predict(input)
y_probs = model.predict_proba(input)[:, 1]
y_preds = (y_probs > 0.3).astype(int)
print(y_preds)
```

Obtendo os resultados previstos, podemos testar se está de acordo com a accuracy do modelo

```python
# Obter informações do modelo registrado
client = mlflow.tracking.MlflowClient()
model_info = client.get_model_version(model_name, model_version)

# Obter o ID do run associado ao modelo
run_id = model_info.run_id

# Recuperar as métricas do modelo
metrics = client.get_run(run_id).data.metrics
accuracy=metrics.get("accuracy")


# Mostrar a accuracy
print("Accuracy do modelo:", accuracy )
```

```python
# Calcular a soma dos valores previstos que há de ser semelhante à accuracy
soma_y_preds = sum(y_preds)
```

```python
# Teste com assert para verificar se a soma dos y_preds faz sentido em relação à accuracy
assert soma_y_preds <= len(y_preds), "Erro: Soma dos y_preds maior que número de previsões"
assert nsamples*accuracy-5*nsamples/10 <= soma_y_preds <= nsamples*accuracy+5*nsamples/10, "Erro: Accuracy fora do intervalo esperado"


print("Testes concluídos com sucesso! 🚀")
```
